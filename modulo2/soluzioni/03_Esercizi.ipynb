1. Usando la funzione `isinstance`, `if/elif/else` e gli operatori booleani (`and`,`or`,`not`), scrivi una funzione con un solo parametro che ritorna:
   - `True` se il parametro è di tipo `bool`;
   - `42` se il parametro è di tipo `int` o `float` e ha un valore inferiore a 10
   - `None` altrimenti


```python
# ES 1

# isinstance(var, tipo)
# ritorna True se variabile è di quel tipo
# False altrimenti

def controlla(var):
    if isinstance(var, bool):
        return True
    elif (isinstance(var, int) or isinstance(var, float)) and var < 10:
        return 42
    else:
        return None

print(controlla(True))
print(controlla(3))
print(controlla(33))
print(controlla([]))
```


3. Data la funzione `flip_coin` che ritorna il risultato del lancio di una moneta, lancia monete in un ciclo while ed esci dal loop (con `break`) solo se hai ottenuto tre `head`.

```python
import random

def flip_coin():
    coin_result = random.randint(0,1)
    if coin_result == 1:
        return "head"
    else:
        return "tail"

# ES 3
count_head = 0
while True:
    result = flip_coin()
    print(result)
    if result == "head":
        count_head += 1
    else:
        count_head = 0
    print(count_head)
    if count_head == 3:
        break
```

5. Scrivi una funzione `homogeneous_list(unnormalized_list)` che restituisce `unnormalized_list` solo se tutti i suoi elementi sono dello stesso tipo, `None` altrimenti. (suggerimento: ricordati che la funzione `type` ritorna il tipo di una variabile)

```python
# ES5
test_list1 = [1, 2, 3, 4]
test_list2 = ["a", "b", "c", "d"]
test_list3 = [{}, "a", []]

def homogenous_list(unnormalized_list):
    first_elem_type = type(unnormalized_list[0])
    for elem in unnormalized_list:
        if not isinstance(elem, first_elem_type):
            return None
    return unnormalized_list

print(homogenous_list(test_list1))
print(homogenous_list(test_list2))
print(homogenous_list(test_list3))
```

```python
test_string = """Python è un linguaggio di programmazione ad alto livello, \
orientato a oggetti, adatto, tra gli altri usi, a sviluppare applicazioni \
distribuite, scripting, computazione numerica e system testing.
Ideato dall'olandese Guido van Rossum all'inizio degli anni novanta, \
è spesso paragonato a Ruby, Tcl, Perl, JavaScript, Visual Basic o Scheme[1].\
Il nome fu scelto per la passione dello stesso inventore verso i Monty Python \
e per la loro serie televisiva Monty Python's Flying Circus[2].\
Spesso è tra i primi linguaggi di programmazione a essere studiati dai neofiti, \
per la sua somiglianza a uno pseudo-codice, e di frequente viene usato \
per simulare la creazione di software grazie alla flessibilità di \
sperimentazione consentita, che permette al programmatore di organizzare\
le idee durante lo sviluppo, come per esempio il creare un gioco tramite\
 Pygame oppure il back-end di un sito web tramite Flask o Django.\
"""



# ES6

def letter_statistics(text):
    letters_dict = {}
    for letter in text:
        if not letter.isalpha(): # skip everything that is not a letter
            continue
        if not letter in letters_dict:
            letters_dict[letter] = 0

        letters_dict[letter] += 1
    return letters_dict

print(letter_statistics(test_string))
```

7. Scrivi una funzione che ritorna `True` se una stringa è palindroma, `False` altrimenti.

```python
def is_palindrome(text):
    reverse_text = ""
    for char in text:
        reverse_text = char + reverse_text
    return reverse_text == text

print(is_palindrome("hello"))
print(is_palindrome("anna"))
```

8. Scrivi una funzione che legge una serie di sequenze di valori separati da virgola (CSV) e restituisce una lista di liste di valori.

```python
example_string = """1,2,3
a,b,c
Idk,some other,string"""

# the result should be [[1,2,3],["a","b","c"],["Idk","some other","string"]]
def parse_csv(csv_string):
    rows = csv_string.split("\n")
    result_list = []
    for row in rows:
        result_list.append(row.split(","))
    return result_list

parse_csv(example_string)
```

9. Scrivi una funzione che prende in input una struttura come quella precedente (lista di liste di stringhe) e la converte in formato CSV.

**BONUS** se riesci, puoi gestire situazioni in cui la stringa contiene "," racchiudendo **solo** queste stringhe fra virgolette " (prova a gestire `evil_list`).

```python
example_list = [[1,2,3],["a","b","c"],["Idk","some other","string"]]
evil_list = [[1,2,3],["a","b","c"],["try converting this one, muhahhaha",
"it has commas, every, where","😈"]]

def naif_csv_converter(row_list):
    csv_string = ""
    for row in row_list:
        row_with_strings = []
        for elem in row:
            row_with_strings.append(str(elem))
        csv_string = csv_string + ",".join(row_with_strings) + "\n"
    return csv_string

print(naif_csv_converter(example_list)) # this works
print(naif_csv_converter(evil_list)) # this doesn't

# BONUS
def better_csv_converter(row_list):
    csv_string = ""
    for row in row_list:
        row_with_strings = []
        for elem in row:
            if isinstance(elem, str) and "," in elem: # add " when a comma is in the string
                elem = f'"{elem}"'
            row_with_strings.append(str(elem))
        csv_string = csv_string + ",".join(row_with_strings) + "\n"
    return csv_string

print(better_csv_converter(evil_list)) # this works
```

10. Scrivi una funzione **ricorsiva** `power(base, exponent)` che calcola l'elevamento a potenza di `base` per `exponent`.
NOTA: l'elevamento a potenza che eleva x all'esponente y è semplicemente x\*x\*x... ripetuto y volte. Qualsiasi numero elevato alla 0 è uguale a 1. 

```python
# ES10
def power_recursive(base, exponent):
    if exponent == 0: # base case
        return 1
    return base * power_recursive(base, exponent - 1) # recursive case

power_recursive(2, 3)
```
