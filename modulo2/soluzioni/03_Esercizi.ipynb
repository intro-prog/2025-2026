1. Usando la funzione `isinstance`, `if/elif/else` e gli operatori booleani (`and`,`or`,`not`), scrivi una funzione con un solo parametro che ritorna:
   - `True` se il parametro è di tipo `bool`;
   - `42` se il parametro è di tipo `int` o `float` e ha un valore inferiore a 10
   - `None` altrimenti


```python
# ES 1

# isinstance(var, tipo)
# ritorna True se variabile è di quel tipo
# False altrimenti

def controlla(var):
    if isinstance(var, bool):
        return True
    elif (isinstance(var, int) or isinstance(var, float)) and var < 10:
        return 42
    else:
        return None

print(controlla(True))
print(controlla(3))
print(controlla(33))
print(controlla([]))
```

```python
import random

def flip_coin():
    coin_result = random.randint(0,1)
    if coin_result == 1:
        return "head"
    else:
        return "tail"

# ES 3
count_head = 0
while True:
    result = flip_coin()
    print(result)
    if result == "head":
        count_head += 1
    else:
        count_head = 0
    print(count_head)
    if count_head == 3:
        break
```

```python
# ES 4
#replace_from_substring("prova", "ov") -> "provit works"

def replace_from_substring(string, substring):
    if substring in string:
        idx = string.find(substring)
        print(idx)

        # usando solo concatenazione
        result = string[:idx] + substring + "it works"

        # usando gli indici
        result = string[:idx + len(substring)] + "it works"
        print(result)

        return result

replace_from_substring("prova", "ov")
```
