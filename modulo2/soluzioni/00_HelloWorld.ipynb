## Markdown

Il markdown è un cosiddetto **linguaggio di markup**. Ciò significa che è un linguaggio che consente di specificare la formattazione di un testo (pensate a word ad esempio). Ad esempio, è possibile rendere testo in **grassetto** scrivendo `**grassetto**`, *corsivo* scrivendo `*corsivo*`. Possiamo anche includere [link](http://wwww.python.org) scrivendo `[link](http://www.python.org)`. Per fare una lista non numerata: 
- capperi
- acciughe

Dobbiamo scrivere:
```
- capperi
- acciughe
```
Per fare una lista numerata:
1. Primo
2. Secondo

Si scrive:
```
1. Primo
2. Secondo
```

Possiamo anche definire titoli e sottotitoli:
# Titolo
## Sottotitolo
### Sotto-sottotitolo

Che in markdown si scrivono così:
```
# Titolo
## Sottotitolo
### Sotto-sottotitolo
```

Ci sono poi molte altre funzionalità, quali immagini, tabelle, etc. Non sono importanti per noi, ma se vi interessa su [markdown.org](https://www.markdownguide.org/) c'è un ottima guida (in inglese).

# Somma tre numeri

Qui sotto c'è una funzione che:
- prende in input tre argomenti
- ne fa la sommma
- ritorna la somma

````python
def somma_tre_numeri(i, j, k):
    print(i + j + k)

print(somma_tre_numeri(2,4,6))
````

````python
# per creare la lista ci sono tre opzioni

# usando un ciclo for
lista_50_numeri = []
for num in range(1,51):
    lista_50_numeri.append(num)

# usando while
lista_50_numeri = []
num = 1
while num < 51:
    lista_50_numeri.append(num)
    num += 1

# usando direttamente range
lista_50_numeri = list(range(1,51))

# usando list comprehensions
lista_50_numeri = [i for i in range(1,51)]


# stampare numeri pari
def stampa_pari(lista):
    for num in lista:
        if num % 2 == 0:
            print(num)
    
stampa_pari(lista_50_numeri)

# trick: uso range per fare già la lista dei numeri pari
print(list(range(2,51,2)))
````
