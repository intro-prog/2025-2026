
2. Crea  un'espressione regolare che estrae anno, mese, giorno di ciascuna data in un testo che contiene date nella forma `aaaa/mm/gg` (dove `a` rappresenta una cifra dell'anno, `m` del mese, `d` del giorno).

```python
# ES3

test_string = "2026/04/22"
regex_date = "([0-9]{4})/([0-9]{2})/([0-9]{2})"

print(re.findall(regex_date, test_string))
match = re.search(regex_date, test_string)
print(f"year is {match.group(1)}")
print(f"month is {match.group(2)}")
print(f"day is {match.group(3)}")
```


3. Data l'espressione regolare precedente, crea una funzione che prende in input una stringa e restituisce una lista di dizionari che contengono i campi `day`, `month` e `year` correttamente popolati usando i valori estratti dall'espressione regolare.

```python
# ES3

def extract_date(text):
    regex_date = "([0-9]{4})/([0-9]{2})/([0-9]{2})"
    result_list = []
    for match in re.finditer(regex_date, text):
        year = match.group(1)
        month = match.group(2)
        day = match.group(3)

        result_list.append({"year": year, 
                            "month": month,
                            "day":day})


    return result_list

test_string = "from 2002/10/23 to 2010/11/12"
print(extract_date(test_string))
```


4. Data l'espressione regolare precedente, crea una funzione che rimpiazza il mese con il mese successivo (attenzione: ci sono solo 12 mesi in un anno)

```python
# ES4
def add_one_month(text): # it only works on a single date in a string
    regex_date = "([0-9]{4})/([0-9]{2})/([0-9]{2})"
    date_match = re.search(regex_date, text)
    month = int(date_match.group(2))
    year = int(date_match.group(1))
    if month == 12:
        month = 1
        year +=1
    else:
        month +=1
    return re.sub(regex_date, rf"{year}/{month:02d}/\3", text)

# WARNING: this does not handle months having a different number of days
print(add_one_month("2025/05/06"))
print(add_one_month("2025/12/06"))
```

5. Uno scrittore sbadato a volte scorda di scrivere una "u" dopo "q", altre volte scrive due "u" dopo "q". Correggi questi errori usando un'espressione regolare

```python
# ES5
def fix_qu_mistakes(text):
    qu_regex = "qu*"
    return re.sub(qu_regex, "qu", text)

print(fix_qu_mistakes("quuuando quali questi"))
print(fix_qu_mistakes("acqa qando"))
```

6. Scrivi una funzione che tokenizza tutte le parole in una stringa

```python
# ES6
def tokenize(string):
    one_word_re = r"\w+"
    return re.findall(one_word_re, string)

print(tokenize("a good day to write some Python"))
```

4. Scrivi una funzione che ritorna un array composto solo dalle righe e dalle colonne di indice dispari di un array bidimensionale (usa slicing). Se l'array ha un numero di dimensioni diverso da due stampa un errore e ritorna `None`

```python
# ES4
# an bidimensional array from 0 to 100 to test
test_array = np.arange(100).reshape((10,10)) 
test_wrong_array = np.zeros((5,3,12))

def even_rows_columns(array):
    if array.ndim != 2:
        print("ERROR: ndim different than 2")
        return None
    return array[1::2,1::2]

print(even_rows_columns(test_array))
print(even_rows_columns(test_wrong_array))
```
