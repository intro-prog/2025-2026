4. Calcola le temperature minime medie giornaliere per estate, inverno, autunno, primavera (usa un ciclo for e indicizzazione avanzata).

6. Quali sono stati i tre giorni di fila con più pioggia? SUGGERIMENTO: puoi usare un ciclo for, iloc e slicing per creare una lista con le precipitazioni cumulative.

```python
# ES6
max_precipitazione = -1
max_idx = -1
for idx in range(len(meteo)):
    three_day_precipitation = meteo.iloc[idx:idx+3]["Precipitazioni (mm)"]
    if three_day_precipitation.sum() > max_precipitazione:
        max_precipitazione = three_day_precipitation.sum()
        max_idx = idx

print(f"la prec massima è {max_precipitazione}")
print(f"la data è {meteo.iloc[max_idx]["Data"]}")
```

```python
# con una lista
precipitazioni_3gg = []
for idx in range(len(meteo)):
    three_day_precipitation = meteo.iloc[idx:idx+3]["Precipitazioni (mm)"]
    precipitazioni_3gg.append(three_day_precipitation.sum())

print(precipitazioni_3gg)
meteo["Precipitazioni 3gg"] = precipitazioni_3gg
meteo.sort_values("Precipitazioni 3gg", ascending=False).iloc[0]
```

7. Crea un DataFrame che contiene le precipitazioni cumulative annue (in termini di mm di pioggia totali). SUGGERIMENTO: puoi usare str.split("-") per separare giorno, mese, anno. Poi puoi usare `str[0]` per accedere all'anno.

```python
# ES7
meteo = pd.read_csv("data/meteo_bologna.csv")
meteo["anno"] = meteo["Data"].str.split("-").str[0]

precipitazioni_per_anno = meteo[["anno", "Precipitazioni (mm)"]].groupby("anno").sum()
precipitazioni_per_anno
```

```python
# ES9
centraline_df = pd.read_csv("data/dati-centraline-bologna-storico.csv", sep=";")

# filter the station and type of pollutant
centraline_df = centraline_df[centraline_df["COD_STAZ"] == "PORTA SAN FELICE"]
centraline_df = centraline_df[centraline_df["AGENTE"] == "PM10"]

# split the date from the time, populate a "Date" column like in the meteo df
centraline_df["Data"] = centraline_df["DATA_INIZIO"].str.split("T").str[0]

# merge with meteo
centraline_meteo_merged = centraline_df.merge(meteo)

# sort by "VALORE" (ascending)
centraline_meteo_merged = centraline_meteo_merged.sort_values("VALORE")
centraline_meteo_merged[["Precipitazioni (mm)", "Stagione"]].iloc[0]
```


10. Calcola per quanti giorni le temperature massime sono state più alte della media per ciascun anno. Ordina il DataFrame risultante per numero di giorni decrescente.

```python
# ES10
meteo = pd.read_csv("data/meteo_bologna.csv")
meteo["anno"] = meteo["Data"].str.split("-").str[0]

# with "as_index=False" I keep the "anno" column
temperature_per_anno = meteo[["anno", "Temperatura media"]].groupby("anno", as_index=False).mean()

temperature_per_anno["Media temperature"] = temperature_per_anno["Temperatura media"]
del temperature_per_anno["Temperatura media"]

# merge the mean temperatures
meteo = meteo.merge(temperature_per_anno)

# add a column to count how many times the max is above the mean
meteo["Maggiore di media"] = meteo["Temperatura massima"] > meteo["Media temperature"]

# summing over a boolean means counting the "True" (which is = 1)
meteo[["anno", "Maggiore di media"]].groupby("anno").sum()
```
