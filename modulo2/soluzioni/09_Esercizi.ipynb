4. Calcola le temperature minime medie giornaliere per estate, inverno, autunno, primavera (usa un ciclo for e indicizzazione avanzata).

```python
# ES4
stagioni = ["Inverno", "Autunno", "Primavera", "Estate"]
for stagione in stagioni:
    temperature_minime_stagione = meteo[meteo["Stagione"] == stagione]
    t_media = temperature_minime_stagione["Temperatura minima"].mean()
    print(f"in {stagione} la temperatura minima media è {t_media}")
```

6. Quali sono stati i tre giorni di fila con più pioggia? SUGGERIMENTO: puoi usare un ciclo for, iloc e slicing per creare una lista con le precipitazioni cumulative.

```python
# ES6
max_precipitazione = -1
max_idx = -1
for idx in range(len(meteo)):
    three_day_precipitation = meteo.iloc[idx:idx+3]["Precipitazioni (mm)"]
    if three_day_precipitation.sum() > max_precipitazione:
        max_precipitazione = three_day_precipitation.sum()
        max_idx = idx

print(f"la prec massima è {max_precipitazione}")
print(f"la data è {meteo.iloc[max_idx]["Data"]}")
```

```python
# con una lista
precipitazioni_3gg = []
for idx in range(len(meteo)):
    three_day_precipitation = meteo.iloc[idx:idx+3]["Precipitazioni (mm)"]
    precipitazioni_3gg.append(three_day_precipitation.sum())

print(precipitazioni_3gg)
meteo["Precipitazioni 3gg"] = precipitazioni_3gg
meteo.sort_values("Precipitazioni 3gg", ascending=False).iloc[0]
```
