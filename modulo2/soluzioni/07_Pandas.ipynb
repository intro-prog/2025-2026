## Cos'è un DataFrame?

Pandas viene utilizzato principalmente per creare delle strutture dati di tipo tablellare, che si chiamano DataFrame. Sono un po' come degli array NumPy bidimensionali, ma ci consentono di fare operazioni molto più sofisticate.

![dataframe pandas](images/pandas_dataframe.png)

Per creare un DataFrame possiamo partire da dizionari e liste Python nel modo seguente:

```python
ghibli_movies = pd.DataFrame({
    "title" : ["Princess Mononoke", "Grave of the Fireflies", "Porco Rosso", "Tales from Earthsea"],
    "director" : ["Hayao Miyazaki", "Isao Takahata", "Hayao Miyazaki", "Gorō Miyazaki"],
    "saddest movie ever?" : [False, True, False, False],
    "year": [1997, 1988, 1992, 2006],
})

ghibli_movies
```

# Indicizzazione di righe e colonne

Come per quanto riguarda dizionari, array, liste, possiamo indicizzare i contenuti dei nostri DataFrame usando le funzionalità appropriate.

Per accedere alle singole righe del nostro DataFrame come faremmo con una lista o un array usiamo l'attributo `iloc`:

```python
print(ghibli_movies.iloc[0])
```

Possiamo anche usare slicing:

```python
ghibli_movies.iloc[1:3]
```

Oltre a poter scorrere le righe del DataFrame, possiamo anche ottenere il contenuto di una o più colonne:

```python
# a single column
print(ghibli_movies["title"])
```

```python
# more than one column
ghibli_movies[["title","year"]]
```

Un'altra modalità di indicizzazione è quella per etichetta o label tramite l'attributo `loc`. Siccome non abbiamo specificato la modalità di indicizzazione del nostro DataFrame, per ora si comporta esattamente come `iloc`.

```python
ghibli_movies.loc[1]
```

Questo comportamento dipende dall'indice (index) del nostro DataFrame. Proviamo a sostituire l'index con gli anni in cui sono stati rilasciati i tre film.

```python
ghibli_movies.index = ghibli_movies["year"]
```

```python
ghibli_movies
```

A questo punto `.loc` si comporta in modo molto diverso!

```python
# this doesn't work
ghibli_movies.loc[0]
```

```python
# this works
ghibli_movies.loc[1997]
```

```python
ghibli_movies.iloc[0]
```

Se non vogliamo avere la colonna `year` ripetuta, possiamo rimuoverla usando `del`

```python
del ghibli_movies["year"]

ghibli_movies
```

Oppure creare il dataframe specificando direttamente la colonna "index":

```python
ghibli_movies = pd.DataFrame({
    "title" : ["Princess Mononoke", "Grave of the Fireflies", "Porco Rosso", "Tales from Earthsea"],
    "director" : ["Hayao Miyazaki", "Isao Takahata", "Hayao Miyazaki", "Gorō Miyazaki"],
    "saddest movie ever?" : [False, True, False, False]},
    index = [1997, 1988, 1992, 2006]
)


ghibli_movies
```

Infine, se vogliamo mantenere il nome della colonna, possiamo assegnarlo usando l'attributo `name` della colonna `index`

```python
ghibli_movies.index.name = "year"
ghibli_movies
```

## Indicizzazione avanzata

Oltre all'indicizzazione "semplice", Pandas ci consente di utilizzare dei filtri più sofisticati. Si tratta di un meccanismo che è analogo a quelli visti con NumPy. In pratica, possiamo selezionare le righe di un DataFrame che rispettano una qualche condizione. Partiamo con un esempio:

```python
# Select movies by Miyazaki

ghibli_movies[ghibli_movies["director"] == "Hayao Miyazaki"]
```

Se una delle colonne contiene valori booleani, possiamo anche filtrare le righe del DataFrame usando direttamente la colonna booleana:

```python
ghibli_movies
```

```python
ghibli_movies[ghibli_movies["saddest movie ever?"]]
```

Possiamo anche definire degli insiemi di valori per filtrare una colonna:

```python
ghibli_movies[ghibli_movies.index.isin([1988,1997])]
```

```python
ghibli_movies[ghibli_movies.index >= 1990]
```

```python
ghibli_movies_noindex = pd.DataFrame({
    "title" : ["Princess Mononoke", "Grave of the Fireflies", "Porco Rosso", "Tales from Earthsea"],
    "director" : ["Hayao Miyazaki", "Isao Takahata", "Hayao Miyazaki", "Gorō Miyazaki"],
    "saddest movie ever?" : [False, True, False, False],
    "year": [1997, 1988, 1992, 2006],
})

ghibli_movies_noindex[ghibli_movies_noindex["year"] >= 1990]
```

## Operazioni su colonne

In modo simile a quanto abbiamo visto per NumPy, possiamo applicare delle operazioni matematiche alle colonne del DataFrame. Prendiamo ad esempio la tabella dei film che abbiamo appena aperto. Per ottenere il numero totale di attori nel film possiamo scrivere:

```python
academyawards["male_cast"] + academyawards["female_cast"]
```

Oltre alle solite operazioni matematiche e booleane, possiamo anche applicare alcune operazioni a colonne composte di stringhe:

```python
# make movie names lowercase
academyawards["movie"].str.lower()
```

```python
# check if a movie name contains "the"
academyawards["movie"].str.contains("the")
```

## Aggiungere righe e colonne

Per aggiungere una riga al DataFrame, possiamo semplicemente usare `loc` per inserire una riga con un indice che non è già presente nel DataFrame

**Attenzione: se abbiamo inserito indici diversi da quelli "sequenziali" dobbiamo usare quelli!**

```python
academyawards.loc[len(academyawards)] = {"movie":"Nomadland",
                                         "imdb": "tt9770150",
                                         "year": 2020,
                                         "award_year": 2020,
                                        "directors": "Chloé Zhao",
                                         "companies": "Highwayman,Hear/Say Productions,Cor Cordium Productions"}

#what happened to the female, male, total cast?
academyawards.tail()
```

```python
# if we specified different indices, we need to use those
ghibli_movies.loc[2001]={"title": "Spirited Away", 
                         "director": "Hayao Miyazaki", 
                         "saddest movie ever?": False}
```

```python
ghibli_movies
```

Per aggiungere una colonna, invece, la aggiungiamo "indicizzando" il suo nome:

```python
# add a column that indicates whether they won the oscar. For now, let's put "False" everywhere
ghibli_movies["won oscar"] = False
```

```python
ghibli_movies
```

Naturalmente, possiamo anche inserire dei valori che sono derivati da altre colonne:

```python
academyawards["total_cast"] = academyawards["male_cast"] + academyawards["female_cast"]
academyawards.tail()
```

## Statistiche

Per quanto riguarda le colonne che contengono valori numerici, Pandas ci offre delle funzionalità per calcolare valori statistici:

```python
# the mean (average) number of male actors in movies
print(academyawards["male_cast"].mean())
```

```python
# the maximum number of total actors in movies
print(academyawards["total_cast"].max())
academyawards[academyawards["total_cast"] == academyawards["total_cast"].max()]
```

```python
# the minimum number of total actors in movies
print(academyawards["female_cast"].min())
```

```python
# some fundamental statistics about the total number of actors
print(academyawards["total_cast"].describe())

#academyawards.tail()
```
