## Stringhe

Quando si trattano dati di tipo linguistico, il tipo di dato più utile è indubbiamente la stringa. Le stringhe sono un tipo di dato **immutabile** e **ordinato** che rappresenta dati testuali di qualche tipo.

Le stringhe in Python sono una sequenza ordinata di caratteri Unicode. Che cosa significa? Significa che possiamo codificare la stragrande maggioranza dei caratteri di quasi qualsiasi sistema di scrittura!


```python
descrizione_totoro = "Il mio vicino Totoro (となりのトトロ, Tonari no Totoro) è un film d'animazione [...]"

print(descrizione_totoro)

emoji_string = "Python strings support emojis as well 🚀🚀🚀"

print(emoji_string)
```

Per indicare le stringhe, vengono tipicamente racchiuse fra virgolette (`"`). Tuttavia è anche possibile usare apicetti singoli (`'`), il che ci consente di utilizzare virgolette al loro interno. Altrimenti, sarebbe necessario usare escape (`\"`)

```python
using_double_quotes = "if i use double quotes I need to escape them as follows: \""
print(using_double_quotes)

using_single_quotes = 'with single quotes i can use them normally: ". However, i need to escape single quotes: \''
print(using_single_quotes)
```

Un'ultima possibilità è quella di creare stringhe di più di una riga: in questo caso possiamo ripetere tre volte le virgolette (`"`).

```python
a_long_string = """this string is so long
I need to split it"""

print(a_long_string)
```


In Python non abbiamo un tipo di dato per i caratteri singoli, tuttavia esse sono costituite da una sequenza ordinata di caratteri, che sono rappresentati come una stringa di lunghezza 1. Ciò significa che possiamo accedere a un carattere specifico della stringa come segue:

```python
# possiamo accedere all'ennesimo carattere della stringa
text = "test"
print(text[0])
```

Come le liste, anche le stringe sono *iterable*, il che significa che possiamo scorrerle un carattere alla volta usando un clico for:

```python
# possiamo applicare un ciclo for sulle stringe (sono iterable)
for character in text:
    print(character)
```

Come per le stringhe, possiamo anche usare lo *slicing*, che ci conente di specificare una "fetta" della stringa usando le parentesi quadre. Lo slicing prende un massimo di tre parametri (`start:stop:step`).

```python
text = "test123"
print(text[0:len(text):2])
```

# Operazioni su stringhe

Come gli interi e i booleani, anche le stringhe supportano una serie di operazioni. In particolare, le stringhe possono essere concatenate con il carattere `+`:

```python
concatenated_string = "idk" + " whatever"
print(concatenated_string)
```

Come per gli interi, posso scrivere `stringa1 += stringa2` per indicare `stringa1 = stringa1 + stringa2`.

```python
my_str = "test"
my_str += "ing"
print(my_str)
```

Un altro operatore è `*`, che ha la forma `stringa * intero` e ripete una stringa tante volte quante viene specificato nell'intero.

```python
print("test" * 3)
```

Come per `+`, posso usare `stringa *= intero` per indicare `stringa = stringa * intero`.

```python
my_str = "test"
my_str *= 3
print(my_str)
```

## Le stringhe sono immutabili

Le stringhe sono un dato **immutabile**, il che significa che possiamo usarle come chiavi nei dizionari, tuttavia significa anche che non possiamo modificarle e che ogni volta che le vogliamo modificare stiamo di fatto creando una nuova stringa!

```python
text = "test"
#text[0] = "b" # questo non funziona, la stringa è immutabile!
text = "b" + text[1:] # questo funziona
text = f"b{text[1:]}"
print(text)

#print(list(text))
```

## Sequenze di escape

Quando scriviamo delle stringhe, è spesso necessario scrivere dei caratteri che non è possibile rappresentare all'interno della stringa stessa, un esempio classico è l'andata a capo (*new line*).
Per questi casi, si usano le cosiddette sequenze di escape. Si tratta di sequenze di caratteri il cui primo è `\` (ad esempio la new line è `\n`).


| Escape Sequence|         Meaning|
|:--------------:|:---------------:|
|             \\\\ |  backslash|
|              \\' |  single quote (dentro a single quote)|
|              \\" |  double quote (dentro a double quote)|
|              \\b |  backspace|
|              \\n |  new line|
|              \\t |  horizontal tab|


```python
print("ciao\nmondo")
print("ciaoo\b")
print("ciao\tmondo")
print("ora voglio fare un backslash \\")
```

## Metodi e Funzioni specifici per stringhe

Oltre alle operazioni che abbiamo già visto, ci sono poi una serie di metodi e funzioni che possono essere applicate alle stringhe. 

### La funzione len

La prima è la funzione `len`, che consente di ottenere la lunghezza (in termini di caratteri) di una stringa:

```python
print(len("test"))
```

### Stringhe e case

Ci sono poi dei metodi che consentono di cambiare il **case** delle stringhe:
- `my_string.lower()` rende tutto il testo minuscolo
- `my_string.upper()` rende tutto il testo maiuscolo
- `my_string.capitalize()` rende la prima lettera maiuscola, tutto il resto minuscolo
- `my_string.title()` rende la prima lettera di ciascuna parola maiuscola, tutto il resto minuscolo

```python
text = "this is a test for case methods. we need UPPERCASE somewhere"
print(text.lower())
print(text.upper())
print(text.capitalize())
print(text.title())
```

### Cercare, contare, rimpiazzare parte di una stringa

Vi sono poi metodi e parole chiavi che consentono di cercare specifiche sotto-stringhe all'interno della stringa ed effettuare varie operazioni:
- `substring in my_string` ritorna `True` se la sottostringa `substring` è contenuta almeno una volta dentro `my_string`
- `my_string.count(substring)` mi consente di contare quante volte la stringa `substring` appare dentro `my_string`. Ritorna 0 se non trova `substring`
- `my_string.find(substring)` consente di trovare dove si trova l'inizio della stringa `substring` dentro `my_string`. Ritorna `-1` se non trova `substring`
- `my_string.replace(substring, replacement)` cerca la stringa `substring` dentro `my_string` e la rimpiazza con `replacement`

```python
# in
text = "this string contains the word \"the\" two times"
print("the" in text)
print("nope" in text)
```

```python
# count
print(text.count("the"))
print(text.count("nope"))
```

```python
# find
print(text.find("the"))
print(text.find("nope"))
```

```python
# replace
print(text.replace("the", "THE"))
print(text.replace("nope", "THE"))
```

### Tipo di contenuto della stringa

Vi sono poi una serie di funzioni che controllano quale sia il contenuto di una stringa. In particolare, possiamo controllare se una stringa `my_string` è totalmente alfabetica (contiene **solo** caratteri di un qualche sistema di scrittura) usando `my_string.isalpha()`

```python
print("hello".isalpha())
print("トトロ".isalpha())
print("some random words".isalpha())
print("abc123".isalpha())
```

Possiamo inoltre di controllare se una stringa `my_string` contiene **solo** caratteri numerici usando `my_string.isnumeric()`.

```python
print("123".isnumeric())
print("123.456".isnumeric())
print("abc123".isnumeric())
```

Infine, è possibile controllare se una stringa è composta **solo** da caratteri di un sistema di scrittura o numerici usando `my_string.isalnum()`

```python
print("123".isalnum())
print("abc123".isalnum())
print("abc 123".isalnum())
```

### Verificare inizio e fine di una stringa

Altri due metodi utili sono `my_string.startswith(substring)` e `my_string.endswith(substring)`. Si tratta di due funzioni che ritornano `True` solo se la stringa`my_string` contiene `substring`, rispettivamente all'inizio e alla fine.


```python
text = "the quick brown fox jumps over the lazy dog"
print(text.startswith("the"))
print(text.startswith("quick"))
```

```python
print(text.endswith("dog"))
print(text.endswith("brown"))

# EXTRA: removeprefix, removesuffix
print(text)
print(text.removeprefix("quick"))
print(text.removesuffix("dog"))
```

### Unire e dividere stringhe

Gli ultimi due metodi che analizzeremo oggi sono forse i più potenti. Partiamo da `my_string.split(substring)`. Questo metodo divide la stringa `my_string` ogni qual volta viene trovata la sotto-stringa `substring`. Il risultato di questa divisione è una lista di stringhe e la sotto-stringa `substring` viene eliminiata.

```python
# split
text = "the quick brown fox jumps over the lazy dog, and lands on the large, wooden table"
print(text.split(","))
print(text.split("the"))
print(text.split("nope"))
```

## Esercizi

1. Crea una funzione che, data una lista di stringhe, calcola la loro lunghezza media (la somma delle lunghezze diviso il numero di stringhe)
2. Crea una funzione `first_or_second()` che prende in input una sottostringa `substring` e due stringhe `s1,s2`. La funzione ritorna 0 se la sottostringa si trova nella prima stringa (`s1`), altrimenti 1 se si trova nella seconda stringa (`s2`) e -1 altrimenti.
3. Crea una funzione `substitute_character(text, index, replacement)` che prende in input una stringa e un indice e va a sostituire il carattere che si trova a quell'indice con `replacement`.

```python
# ES1

def average_string_length(string_list):
    number_strings = len(string_list)
    total_length = 0
    for s in string_list:
        total_length += len(s)
    return total_length / number_strings

text_list = ["a", "random", "assortment", "of", "strings"]
print(average_string_length(text_list))
```

```python
# ES2
def first_or_second(substr, s1, s2):
    if substr in s1:
        return 0
    elif substr in s2:
        return 1
    else:
        return -1
    

print(first_or_second("a", "ciao", "mondo"))
```

```python
# ES3

def substitute_character(text, index, replacement):
    first_part = text[:index]
    second_part = text[index + 1:]
    return first_part + replacement + second_part

print(substitute_character("ciao", 2, "o"))

str_list = list("ciao")
str_list[2] = "o"
print("".join(str_list))
```
