## Comprehensions

Per list/dict/set/tuple comprehensions si intende una scorciatoia che può essere usata in Python per creare rapidamente strutture dati. Partiamo considerando le liste (*list comprehension*).

La sintassi usata in questo caso è:
```python
 mia_lista = [espressione for elemento in iterabile] 
```

Con questo codice chiediamo a Python di creare una lista che contiene il risultato di *espressione* per ciascun *elemento* contenuto in *iterabile*
Ma cosa significa in questo caso iterabile? Indica qualcosa che possiamo "scorrere" usando un iterazione (ad esemio un ciclo for).

L'espressione precedente è una versione più compatta di:
```python
mia_lista = []
for elemento in iterabile:
    mia_lista.append(espressione)
```

In pratica, possiamo scrivere una espressione come la seguente:


````python
my_list = [a*2 for a in range(5)]
print(my_list)
````

**Prima di eseguirlo** secondo voi che cosa produce questo codice?

Oltre a questa forma più banale della list comprehension, possiamo anche inserire delle condizioni al suo interno, usando la seguente sintassi:

```python
 mia_lista = [espressione for elemento in iterabile if condizione] 
```

In questo caso creiamo una lista, che contiene il risultato di *espressione* per ciascun *elemento* di *iterabile* se è vera la *condizione*. In pratica, è equivalente a scrivere:

```python
mia_lista = []
for elemento in iterabile:
    if condizione:
        mia_lista.append(espressione)
```

**Domanda**: di che tipo è la *condizione*? È lo stesso tipo di dato adottato per le espressioni condizionali (if/elif) e per i cicli while.

Vediamo come applicare questa struttura in un esempio pratico:




````python
lista_frutta = ["banana", "arancia", "pompelmo", "ananas", "mela"]

lista_filtrata = [frutto for frutto in lista_frutta if "o" in frutto]
print(lista_filtrata)
````

Che cosa produce questa espressione?

**NB** le comprehensions sono certamente utili, ma hanno un grande svantaggio: se scriviamo codice molto complicato al loro interno diventa sempre più difficile capire cosa fanno. Il mio consiglio è di evitare di usare `if` dentro le list comprehensions, preferendo invece un ciclo for. Ho volutamente evitato di mostrarvi anche un modo per inserire `else` dentro alle comprehensions, perché non è quasi mai utile.

## Tuple, Dict Comprehensions

Oltre alle comprehensions su liste, possiamo anche scrivere comprehensions per `tuple` e per `dict`. Vediamo subito come.

Le tuple comprehension si comportano esattamente come le liste.

````python
# tuple comprehension

my_tuple = (frutto for frutto in lista_frutta if "o" in frutto)
print(tuple(my_tuple))
````

Per i dizionari dobbiamo fornire una chiave e un valore, ma il comportamento è sempre molto simile:

````python

# dict comprehension

my_dict = {frutto:len(frutto) for frutto in lista_frutta}
print(my_dict)
````

## Iterazione

Avete già visto i due principali cicli che ci consentono di ripetere (iterare) una operazione più volte. Si tratta di *for* e *while*. Tuttavia è utile discutere alcuni aspetti più pratici di questi costrutti. Partiamo da istruzioni che controllano il flusso di esecuzione dei cicli, `break` e continue.

### Break e continue

Il costrutto `break` è forse il più intuitivo: quando viene invocato, viene immediatamente interrotto il ciclo.
Consideriamo ad esempio il codice seguente:



````python
i=0

while i<10:
    print(i)
    i += 1
````

Potrebbe essere riscritto come:

````python
i = 0
while True:
    if i == 10:
        break
    print(i)
    i += 1
````


In pratica, sto imponendo al ciclo di terminare quando si verifica una data condizione.

`Continue` è invece l'opposto: l'esecuzione del ciclo prosegue ignorando tutto il resto del codice. Consideriamo un altro esempio:

````python
test_list = [0,1,2,3,4]
for num in test_list:
    if not(num % 3 == 0):
        print(num)
````

````python
test_list = [0,1,2,3,4]
for num in test_list:
    if num % 3 == 0:
        continue
    print(num)
````

### Quando usare break/continue

A volte durante la scrittura del codice può succedere di dover inserire condizioni tramite if/elif/else che iniziano a contenere una gran parte del codice presente dentro ai cicli. In questo caso, è spesso utile usare `continue` e `break` per evitare di dover mettere tutto il codice dentro a costrutti if/elif/else.

## Cicli for in Python

I cicli for di Python ci consentono di semplificarci molto la vita quando trattiamo strutture dati annidate. In alcune situazioni, potremmo voler sfruttare una lista che contiene al suo interno liste (o tuple) di due elementi. In questo caso, è spesso utile ottenere due variabili che contengono questi due elementi, invece di doverli assegnare manualmente. Immaginiamo di avere una lista che contiene nomi di studenti e la loro matricola:


````python
a_nested_list = [("Michele Corazza", 429918), ("Fabio Le Piane", 611203), 
["Pippo", 0000]]
````

Per ottenere nome e matricola di ciascuno studente, possiamo scrivere:

````python
for student_info in a_nested_list:
    student_name = student_info[0]
    student_number = student_info[1]

    print(f"{student_name} ha matricola {student_number}")
````

Per semplificare un po' il codice, è possibile usare una scorciatoia, che ci consente di assegnare due variabili alla volta:

````python
for student_info in a_nested_list:
    print(f"student_info ha tipo {type(student_info)}, lunghezza {len(student_info)}")
    student_name, student_number = student_info
    print(f"{student_name} ha matricola {student_number}")
````

Ma Python ci aiuta ulteriormente! Possiamo assegnare le due variabili direttamente nel ciclo for:

````python
for student_name, student_number in a_nested_list:
    print(f"{student_name} ha matricola {student_number}")
````

Questo risulta particolarmente utile quando si usa la funzione `enumerate`, che prende in input un iterabile e restituisce un iterabile che al suo interno contiene una tupla costituita da un indice crescente e da ciascun elemento dell'iterabile.

````python
fruit_list = ["apple", "orange", "grapefruit"]

print(list(enumerate(fruit_list)))
````

Questo si adatta molto bene all'utilizzo dentro i cicli for, perché possiamo assegnare sia l'indice che il contenuto della lista in una sola riga.

````python
for idx, fruit in enumerate(fruit_list):
    print(f"{fruit} has index {idx}")
````

## Iterare dizionari

L'ultimo argomento di oggi riguarda l'iterazione di dizionari. Che cosa succede se applichiamo un ciclo for direttamente al dizionario?

````python
fruit_to_color = {}
fruit_to_color["apple"] = "green or red"
fruit_to_color["orange"] = "orange"

print(fruit_to_color["apple"])
````

````python
fruit_to_color = {"orange": "orange", 
                  "apple": "green or red",
                  "pineapple": "not on pizza"}

for fruit in fruit_to_color:
    print(fruit)
````

Vengono ritornate le **chiavi**, questo è equivalente a invocare il metodo `.keys()` sul dizionario.
Se invece volessimo ottenere i valori, potremmo usare il metodo `.values()` come segue:

````python
for color in fruit_to_color.values():
    print(color)
````

L'ultimo caso riguarda la situazione nella quale volessimo ottenere sia le chiavi che i valori in un ciclo for. In questo caso usiamo il metodo `.items()` e la scorciatoia dei clili for che vi ho appena mostrato per stamparli entrambi.

````python
for fruit in fruit_to_color:
    color = fruit_to_color[fruit]
    print(f"{fruit} has color {color}")

for fruit, color in fruit_to_color.items():
    print(f"{fruit} has color {color}")
````
