## La nostra prima figura

Prima di iniziare, importiamo `matplotlib.pyplot`. Come per numpy e pandas, c'è un modo "canonico" di importare matplotlib, che è il seguente:

```python
import matplotlib.pyplot as plt
import numpy as np # anche numpy ci sarà utile, importiamolo
import pandas as pd # e anche pandas
```

Matplotlib crea grafici su una `Figure` che contiene uno o più `Axes`. Se la `Figure` è l'intera "finestra" in cui andiamo a creare i grafici, gli `Axes` sono parti della figura in cui possiamo usare coordinate bidimensionali.

Partiamo quindi disegnando la figura più semplice a cui possiamo pensare, una linea:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,3,4])
plt.show()
```

Che cosa abbiamo appena fatto? Abbiamo creato una figura usando la funzione `subplots`, che ci restituisce una figura e un asse. A partire dall'asse abbiamo poi invocato il metodo `plot` su due argomenti di tipo lista, che in questo caso contengono entrambe i numeri tra 1 e 4. Questi due argomenti rappresentano le coordinate dei punti sull'asse delle x e sull'asse delle y. In questo caso abbiamo disegnato una linea che attraversa tutti i punti con coordinate (1,1),(2,2),(3,3),(4,4).
Infine, abbiamo scritto `plt.show()` per visualizzare la figura.

Prima di addentrarci nelle varie funzionalità, iniziamo da una visione di insieme di tutte le parti di un grafico matplotlib:

![matplotlib anatomy](images/matplotlib_anatomy.webp)

Cominciamo quindi a cambiare alcuni degli attributi della nostra figura.

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,3,4])
ax.set_xlabel("this is pointless")
ax.set_ylabel("this is also pointless")
ax.set_title("our first figure")
plt.show()
```

## Personalizzare il grafico

Vediamo ora come alterare il grafico della nostra linea. Ad esempio, potremmo voler visualizzare i punti che stiamo disegnando con dei cerchi. In questo caso, possiamo scrivere:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,3,4], "o")
plt.show()
```

Per disegnare i cerchi abbiamo utilizzato una sintassi particolare che nella documentazione viene chiamata `fmt`, che sta per "format string". Si tratta di un modo più conciso di definire il comportamento della nostra curva. In questo caso la stringa "o" rappresenta dei pallini. Se volessimo invece fare delle croci rosse, potremmo scrivere:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,3,4], "r+")
plt.show()
```

Se volessimo specificare in modo esplicito le proprietà della nostra curva, potremmo scrivere:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4], [1,2,3,4], color='green', marker='o', linestyle='dashed',
     linewidth=2, markersize=12) # una linea tratteggiata verde, con pallini in corrispondenza dei punti
plt.show()
```

E se volessimo disegnare più di una singola curva? Possiamo farlo invocando "plot" più volte:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,3,4], "o-", label="increasing")
ax.plot([1,2,3,4],[4,3,2,1], "+-", label="decreasing")
ax.legend()
plt.show()
```

## Matplotlib e Numpy

Oltre a liste, matplotlib supporta anche l'utilizzo di array numpy.

```python
fig, ax = plt.subplots()
x = np.arange(1,5)
y = np.arange(1,5)
ax.plot(x,y, "o-")
plt.show()
```

## Legenda e label

Per ottenere una legenda, dobbiamo prima assegnare una label a ciascuna curva che stiamo disegnando e invocare il metodo `legend` sugli assi. Di default, verranno mostrati i colori e le label delle nostre due curve.

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,3,4], "o-", label="increasing")
ax.plot([1,2,3,4],[4,3,2,1], "+-", label="decreasing")
ax.legend()
plt.show()
```

## Pie chart

Per ora abbiamo visto solo grafici formati da una o più linee. Tuttavia matplotlib può essere usato per disegnare molti tipi di grafico diversi. Un primo esempio è il grafico a torta o pie chart. Questo grafico è molto utile per visualizzare dati numerici e comparare la loro grandezza. Ha senso usare un grafico a torta **solo quando la somma di tutti i valori ha un qualche senso**.

Vediamo come creare pie chart con `matplotlib`:

```python
counts = [10, 35, 8, 66]
categories = ["18-25", "25-30", "30-40", "40+"]

fig, ax = plt.subplots()
ax.pie(counts, labels=categories)
plt.show()
```

Possiamo anche specificare i colori per ogni "fetta":

```python
counts = [10, 35, 8, 66]
categories = ["18-25", "25-30", "30-40", "40+"]
colors = ["green", "purple", "black", "orange"]

fig, ax = plt.subplots()
ax.pie(counts, labels=categories, colors=colors)
plt.show()
```

Possiamo anche mostrare la percentuale di ciascuna fetta usando il parametro `autopct`, che viene di solito usato per la percentuale.

```python
counts = [10, 35, 8, 66]
categories = ["18-25", "25-30", "30-40", "40+"]

fig, ax = plt.subplots()
ax.pie(counts, labels=categories, autopct="%.2f")
plt.show()
```

## Bar chart

Le pie chart sono utili, tuttavia hanno anche una serie di limiti non secondari. Uno dei problemi, in particolare, è che non è sempre facile comparare i valori di fette che non differiscono di molto. Per questa ed altre situazioni (ad esempio quelle in cui la somma totale dei valori non è un concetto sensato) si usano i bar chart. Vediamo come farli in matplotlib.

```python
x = [1,2,3]
y = [10, 8, 6]

fig, ax = plt.subplots()
ax.bar(x,y)
plt.show()
```

In questo caso x indica il punto in cui la barra verrà messa nel grafico. Solitamente però i bar chart sono usati con dati categorici, non numerici. In questo caso possiamo scrivere:

```python
counts = [10, 35, 8, 66]
categories = ["18-25", "25-30", "30-40", "40+"]

fig, ax = plt.subplots()
ax.bar(categories, counts)
plt.show()
```

E se volessimo mostrare più barre nello stesso grafico? Purtroppo matplotlib non ci aiuta molto e diventa un po' più complicato. Dobbiamo infatti posizionare manualmente tutte le barre al punto giusto, gestendone la posizione. Visto che siamo in giornate di tumulto per i dazi, vediamo qualche dato su import e export di vari stati:

```python
countries = ["Italy", "France", "Germany", "China", "USA"]
import_export_values = {
    "import" : [640, 786, 1476, 2557, 3172],
    "export" : [677, 649, 1718, 3379, 2021]
}
x = np.arange(len(countries))  # la posizione delle barre per ogni paese è un numero intero da 0 al numero di paesi
width = 0.25  # la larghezza delle barre
multiplier = 0 # un numero che verrà moltiplicato per l'offset per ottenere la posizione della barra corretta

fig, ax = plt.subplots()

for label,value in import_export_values.items():
    offset = width * multiplier # la posizione del paese corrente
    rects = ax.bar(x + offset, value, width, label=label)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Billions of dollars')
ax.set_title('Countries')
ax.set_xticks(x + width, countries) # imposta i nomi dei paesi in basso
ax.legend(loc='upper left', ncols=3) # mostra la legenda

plt.show()
```

## Scatter plot

Un altro tipo di grafico è lo scatterplot. Si tratta di un meccanismo per visualizzare dati bidimensionali in una griglia. Può essere utile per vedere se c'è una relazione fra due valori. Un esempio classico (che vi mostro ora) è la relazione fra altezza e peso delle persone.

```python
import pandas as pd

data = pd.read_csv("data/weight_height.csv")
weight = data["Weight"]
height = data["Height"]

fig, ax = plt.subplots()
ax.scatter(weight, height)
plt.show()
```

E se volessimo dare colori diversi ad alcuni punti? Nel dataset che ho trovato non ci sono distinzioni di genere, ma per farvi vedere creerò una colonna sintetica (casuale) nel nostro dataframe pandas.

```python
data["sex"] = np.random.randint(0,2, size=len(height))

fig, ax = plt.subplots()
ax.scatter(weight, height)
plt.show()
```

E se volessimo mettere una legenda? Vediamo come fare.

```python
fig, ax = plt.subplots()
for i in range(2): # una volta per ciascun genere
    curr_table = data[data["sex"] == i]
    ax.scatter(curr_table["Weight"], curr_table["Height"])
plt.show()
```

## Box and whiskers plot

Come ultima tipologia di grafico (ce ne sarebbero altre decine) vediamo il **diagramma a scatola e baffi** (in inglese box and whiskers plot). Si tratta di una tipologia di grafico che ci consente di osservare in una sola immagine l'andamento di una lunga serie di dati. In particolare, vengono mostrati una serie di valori:
- **mediana**: il valore "di mezzo" della distribuzione, il 50% dei numeri è più basso di questo valore e il 50% è più alto
- **primo e quarto quartile**: il primo quartile è il valore per cui il 25% della serie ha valore più basso. Il terzo quartile è quello per cui il 25% della serie ha valori più alti
- **minimo e massimo** i punti più bassi e più alti del dataset **escludendo dati anomali** (outliers)

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

penguins = sns.load_dataset("penguins")

penguins
```

```python
# Draw a nested barplot by species and sex
g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
plt.show()
```
