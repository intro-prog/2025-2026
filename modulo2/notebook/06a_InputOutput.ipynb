# File Input/Output

Quando vogliamo fare operazioni più complesse è spesso utile/necessario caricare e salvare dati che sono memorizzati su file del nostro computer. Oltretutto, se i nostri calcoli richiedono molto tempo questo ci obbliga a salvare i risultati: altrimenti ogni volta dovremo rifare la stessa (lunga) procedura.

Il codice qui sotto vi mostra come usare la funzione `open()`, che serve ad aprire i file. Usando poi la funzione `read()` possiamo leggere **tutto il contenuto** del file.

Notate come una volta finito di leggere il file dobbiamo chiuderlo. Questo è specialmente importante quando dobbiamo scrivere un file.

Vedremo anche un modo più semplice per gestire la chiusura automatica di un file fra poco.

```python
infile = open('data/small_file.txt', 'r')

print(infile.read())

infile.close()
```

La funzione `open()` apre un file e ci ritorna un **file object** (prova a stampare il tipo del file). I parametri che richiede sono:
- **filename**: il percorso del file da aprire. Possiamo usare un percorso **assoluto** o **relativo**.

- **mode**: la modalità con cui apriamo il file. I valori più comuni sono `r` (read) per leggere un file (è il valore di default), `w` (write) per scrivere un file (sovrascrive il file se esiste) e `a` (append) per concatenare qualcosa a un file già esistente.

Per altre informazioni su open puoi guardare la [documentazione](https://docs.python.org/3/library/functions.html#open).

>**IMPORTANTE**: ciascun file aperto va chiuso, o potrebbe non essere disponibile per questo o altri programmi!

Ci sono varie maniere per leggere un file. Abbiamo visto `read()`. Possiamo specificare un numero di caratteri da leggere usando la seguente sintassi:

```python
infile = open("data/small_file.txt")
print(infile.read(3))
infile.close()
```

Ogni volta che leggiamo un pezzo di un file e lo lasciamo aperto continuiamo a leggere da dove avevamo finito. Possiamo ricominciare usando `.seek(0)` sul file object.

```python
infile = open("data/small_file.txt")

print(infile.read(3))
print("continue reading")
print(infile.read(3)) # it continues after the first three characters

infile.seek(0)
print("reset the read")
print(infile.read(3))
infile.close()
```

Un'altra funzione utile di python è quella che ci consente di **iterare** le righe di un file. Usando un ciclo for sul file object, ciascuna riga è un elemento di una sequenza:

```python
#is something weird happening in the print? why is that?
infile = open("data/small_file.txt")
for line in infile:
    print(line)

infile.close()
```

Possiamo anche leggere tutte le righe e trasformarle in una lista:

```python
infile = open("data/small_file.txt")
line_list = infile.readlines()
infile.close()
print(line_list)
```

## Aprire file con With

Python fornisce anche `with`, che ci consente di delimitare un blocco di codice.

Nel contesto dei file, usare `with` ha vari vantaggi.
- non è necessario chiudere un file, lo fa da solo quando esce da `with`
- anche se il programma dovesse andare in crash, il file verrebbe chiuso lo stesso.
- il codice è più pulito

Quando possibile, meglio usare `with` per evitare problemi. Per leggere le righe di un file si può quindi scrivere così:

```python
with open('data/small_file.txt') as infile:  # The file is opened
    
    lines = infile.readlines()
    
# As soon as we exit the indented scope, the file is closed again 
# (and made available to other programs on your computer)
print(lines)
```

Il codice nel blocco indentato (dentro `with`) viene eseguito mentre il file viene aperto. Il file viene chiuso automaticamente quando il blocco è chiuso.

## Scrivere file

Per scrivere un file in python possiamo usare una sintassi simila a quella di `open()` e `read()` che abbiamo appena visto. Le differenze sono:
- dobbiamo specificare il **mode** come `w`
- dobbiamo usare `write()` per scrivere sul file

> **Attenzione!** Aprire un file esistente con mode `w` lo svuota completamente!

```python
# The folder you with to write the file to ('output' below) has to exist on the file system

with open('output/output-test-1.txt', 'w') as outfile:
    outfile.write("My name is:")
    outfile.write("John")
```

Quando si scrive riga per riga è **nostro compito** gestire le andate a capo usando la sequenza di escape per newline `\n`!

```python
with open('output/output-test-2.txt', 'w') as outfile:
    
    outfile.write("My name is:\n")
    outfile.write("Alexander")
    
    
    outfile.write("ééèèüAæøå\n")
    outfile.write("平仮名")
    
```

## Looping through folders and files

Se volessimo aprire tutti i file di una cartella, possiamo usare due funzioni del pacchetto `os`: `os.listdir()` e `os.walk()`. 

`os.listdir()` usa come input una cartella e restituisce semplicemente una lista dei contenuti in quella cartella. `os.path.join()` viene usato per unire il nome del file con la cartella a cui appartiene, mentre `os.path.isfile()` controlla se il percorso specificato indica un file.

```python
dir_name = "test/programmazione"
file_name = "note.txt"
print(os.path.join(dir_name,file_name))
```

```python
import os  # package with operating system functions
```

```python
folder_path = "data/multiple-files"

print(os.listdir(folder_path)) #check the folder: see anything interesting?
```

```python
#listdir can be iterated over like a list
for file_name in os.listdir(folder_path):
    
    file_path = os.path.join(folder_path, file_name)
    
    if not os.path.isfile(file_path):  # Why this?
        continue
    
    print("i found file " + file_name)
    
    with open(file_path) as infile:
        print("it has the following content")
        print(infile.read())
```

La funzione `os.walk()` è utile per esplorare un intero **sottoalbero** contenuto in una data cartella, ovvero tutti i file e le cartelle che sono contenute nella cartella. Ad esempio, immaginiamo di avere una cartella "documenti" che contiene una cartella "casa" e un file "bollette.txt" all'interno di "casa". `os.listdir()` mi mostrerà solo la cartella "casa", mentre `os.walk` esplorerà anche i suoi contenuti e il file "bollette.txt". `os.walk()` ritorna una lista di tuple, che contengono la cartella che il modello sta esplorando (`root`), le sottocartelle che essa contiene (`dirs`) e i file e le cartelle contenuti al suo interno (`files`).

```python
folder_path = "data/multiple-files"

list(os.walk(folder_path))
```

```python
# to loop over files found by os.walk, we first loop over its results
# then we do a for loop over the "files"
for root, dirs, files in os.walk("data/multiple-files"):
    for file_name in files:
    
        file_path = os.path.join(root, file_name)
            
        print("i found file " + file_name)
    
        with open(file_path) as infile:
            print("it has the following content")
            print(infile.read())
```
