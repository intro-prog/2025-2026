# Espressioni Regolari

Le **espressioni regolari** (*Regular Expression* in inglese, *regex* abbreviato) sono uno strumento che ci consente di specificare un insieme di stringhe tramite una stringa. Si tratta di uno strumento molto potente, che consente di:
- Trovare occorrenze (*match*) della regular expression all'interno di una stringa;
- Estrarre parti di una stringa che sono circondate da testo che ha determinate caratteristiche;
- Effettuare ricerca/sostituzione su testi in maniera sofisticata.

In Python, le regular expression sono implementate in un pacchetto chiamato `re`, che contiene tutte le funzionalità necessarie per utilizzarle.

Sebbene la sintassi delle regex abbia delle specificità per il singolo strumento o linguaggio di programmazione, la gran parte della sintassi è condivisa. Questo significa che una volta imparate le regular expression in Python sarete in grado di usarle anche in altri tool.

## Regex in Python

Le espressioni regolari in Python ci consentono di utilizzare una serie di caratteri speciali che hanno degli effetti ben precisi sull'espressione regolare. Cominciamo a capire come funzionano le espressioni regolari da alcuni semplici esempi. La documentazione ufficiale del pacchetto `re` contenente tutta la sintassi e le funzioni implementate si trova qui: https://docs.python.org/3/library/re.html

Uno strumento particolarmente utile per scrivere regex è https://regex101.com, che consente di controllare cosa faccia un espressione regolare e testarla. Per usarlo dovete selezionare "Python" dal menu a sinistra.

Cominciamo importando il pacchetto `re`

````python
import re
````

Per manipolare regex dobbiamo usare delle funzioni specifiche del pacchetto `re`. In particolare, le funzioni più utili sono:
- `re.match(pattern, string)` ci consente di determinare se uno o più caratteri **all'inizio della stringa** `string` fanno *match* con l'espressione regolare `pattern`. Ritorna un oggetto `Match` se trova qualcosa, `None` altrimenti;
- `re.search(pattern, string)` cerca la prima occorrenza che fa *match* con `pattern`, non per forza all'inizio della stringa `string`. Ritorna un oggetto `Match` se trova qualcosa, `None` altrimenti.
- `re.findall(pattern, string)` cerca **tutte** le occorrenze che fanno *match* con `pattern` in `string`. Ritorna una lista di stringhe (o una lista di tuple contenenti stringhe).
- `re.finditer(pattern,string)` cerca **tutte** le occorrenze che fanno *match* con `pattern` in `string`. Ritorna un iteratore (un oggetto che "sputa" fuori output uno alla volta) di `Match`.
- `re.sub(pattern, replacement, string)` cerca tutte le occorrenze di `pattern` in `string`, rimpiazzandole con `replacement`

Possiamo ora iniziare a lavorare sulla forma più semplice di espressione regolare. Possiamo infatti scrivere una stringa "semplice" e trovare tutte le sue occorrenze all'interno di un'altra stringa.

````python
simple_regex = "is"
simple_string = "My house is in Bologna. It is in the city center"

# search the first occurrence of "is"
print(re.search(simple_regex, simple_string))

# print(re.match(simple_regex, simple_string)) # what would this output?
# print(re.findall(simple_regex, simple_string)) # and this?
# print(list(re.finditer(simple_regex, simple_string))) # this is an iterator
````

### Character set

I primi caratteri speciali che vediamo oggi sono le parentesi quadre `[]`. Questi delimitano un insieme di caratteri che consideriamo come validi. Ad esempio, se scriviamo `[ab]` la nostra regex indica stringhe di un carattere, che può essere a o b.

````python
simple_regex = "[ab]"
simple_string = "best airport"

print(re.findall(simple_regex, simple_string))
````

**ATTENZIONE** le maiuscole e le minuscole contano!

Oltre a scrivere manualmente uno o più caratteri, si possono anche esprimere "intervalli" di caratteri utilizzando un `-`. Per esempio `[a-d]` fara match con tutti i caratteri a,b,c,d. Allo stesso modo `[0-5]` farà match di tutti i caratteri da 0 a 5. 

**ATTENZIONE**, si tratta di caratteri, non di numeri! Non posso scrivere `[0-1000]` per ottenere i numeri fra 0 e 1000, devo gestire ogni carattere in modo diverso.

````python
simple_regex = "[A-C]"
simple_string = "BCAZH"

print(re.findall(simple_regex, simple_string))
````

````python
# i can also use multiple ranges!

simple_regex = "[A-Z0-4]"
simple_string = "0BCAZH09"

print(re.findall(simple_regex, simple_string))
````

````python
#A-z denotes any upper or lowercase character

simple_regex = "[A-z]"
simple_string = "this is a TEST"

print(re.findall(simple_regex, simple_string))
````

### Cardinalità

Nelle espressioni regolari Python c'è una serie di caratteri che viene usata per indicare quante volte appare l'espressione precedente, fra cui:
- `*`: indica che l'espressione precedente appare zero o più volte
- `+`: indica che l'espressione precedente appare una o più volte
- `?`: indica che l'espressione precedente appare zero o una volta
- `{m}`: indica che l'espressione precedente appare esattamente `m` volte

Facciamo qualche esempio:

````python
simple_regex = "a+b"
simple_string = "aaaaabc"

print(re.search(simple_regex, simple_string))
````

````python
simple_regex = "[0-9]*B"
simple_string = "099BCA"

print(re.search(simple_regex, simple_string))
````

````python
simple_regex = "[0-9]*B"
simple_string = "BCA"

print(re.search(simple_regex, simple_string)) #what happens here?
````

````python
simple_regex = "[0-9]?B"
simple_string = "099BCA"

print(re.search(simple_regex, simple_string)) # what happens here?
````

````python

simple_regex = "[0-9]{2}B"
simple_string = "099BCA"

print(re.search(simple_regex, simple_string)) # what happens here?
````

### Caratteri speciali

Oltre ai gruppi e ai caratteri "semplici", ci sono una serie di caratteri che hanno un ruolo particolare nelle regex:
- `.` il punto rappresenta qualsiasi carattere;
- `^` rappresenta l'inizio della stringa;
- `$` rappresenta la fine della stringa

***ATTENZIONE*** `^` dentro a un insieme di caratteri indica stringhe non comprese nella lista di caratteri successiva

````python
simple_regex = "B.A"
simple_string = "BCADA"

print(re.findall(simple_regex, simple_string))
````

````python
simple_regex = "B.+A"
simple_string = "BCADA"

print(re.findall(simple_regex, simple_string))
````

````python
date = "01/10/1998"
day_regex = "^[0-9]+"

print(re.findall(day_regex, date))
````

````python
date = "01/10/1998"
year_regex = "[0-9]+$"

print(re.findall(year_regex, date))
````

### Raw strings e regex

Come abbiamo discusso quando abbiamo parlato di stringhe, ricordatevi che `\` viene usato per creare le sequenze di escape.

Alcuni caratteri speciali delle regex usano `\`, per cui dovremmo scrivere ad esempio `\\w` al posto di `\w` per evitare che `\` sia interpretato come un carattere di escape.
Per fortuna Python ci consente di usare le cosiddette "raw string", che trattano "\" come un carattere normale, ignorando quindi il suo ruolo per fare sequenze di escape. Per scriverle si usa `r` prima delle virgolette.

Nota: le versioni moderne di python risolvono "da sole" questo problema in molti casi, ma è comunque buona prassi usare raw string quando usiamo regex per evitare problemi.

````python
newline = "\n"
raw_string = r"\n"

# how long are these two sequences?
print(len(newline))
print(len(raw_string))
````

### Sequenze speciali

Oltre ai caratteri speciali, abbiamo anche una serie di sequenze speciali. Ce ne sono tante, fra le quali:
- `\w` indica qualsiasi carattere indicato come "word character" (lettere da a a z, numeri da 0 a 9, il carattere `_`, caratteri in altri sistemi di scrittura)
- `\b` indica un carattere di "confine di parola". Può venire usato per estrarre una parola evitando i casi in cui è contenuta in altre
- `\d` indica un carattere numerico
- `\s` indica qualsiasi carattere di spaziatura (spazi, tab, etc)

````python
testo = "Il mio vicino Totoro\t(となりのトトロ?, Tonari no Totoro) è un film del 1988"

# \w = lettere, numeri, underscore
parole = re.findall(r"\w+", testo)
print("Sequenze \\w:", parole)
````

````python
text_word_boundary = "bat rat at"

# \b = confine di parola
at_as_word = re.findall(r"\bat\b", text_word_boundary)
print(at_as_word)

at_alone = re.findall(r"at", text_word_boundary)
print(at_alone)
````

````python
# \d = cifre numeriche
numeri = re.findall(r"\d", testo)
print("Cifre \\d:", numeri)
````

````python
# \s = spazi, tab, newline, ecc.
spazi = re.findall(r"\s", testo)
print("Caratteri \\s:", spazi)
print("Numero di spaziature trovate:", len(spazi))
````

### Gruppi

Un ulteriore funzionalità del pacchetto `re` riguarda i cosiddetti gruppi. Vengono denotati con le parentesi tonde `()` e contengono una sequenza di caratteri da considerare in maniera separata. Questo può essere utile per varie ragioni, una di esse è il fatto che ci consentono di fare match di un gruppo di caratteri che si ripetono.

````python
test_string = "lalala l a l a"
test_regex = r"(la)+"

print(re.search(test_regex, test_string))
````

Tuttavia, la funzionalità principale di questi gruppi è che sono **capturing**, il che significa che vengono memorizzati e possono essere usati per le sostituzioni! Data una stringa della seguente forma:

```python
"(text) (text) (text)"
```
Accediamo ai contenuti del primo gruppo con `r"\1"`, del secondo con `r"\2"` e così via:

````python
test_string = "the beautiful mountain"
test_regex = r"(\w+) (\w+) (\w+)"

print(re.sub(test_regex, r"\1 ugly \3",test_string))

print(re.sub(test_regex, r"\1 \2 coast",test_string)) # what does it print?
````

Alcune funzioni (come `findall`) ritornano l'ultimo gruppo trovato quando un gruppo si ripete. Per questa ragione, è spesso utile creare gruppi **non-capturing**, il cui contenuto quindi non viene mantenuto in memoria e non può essere utilizzato nelle sostituzioni.

Questi gruppi vengono scritti con `(?:text)`.

````python
test_string = "lalala l a l a"
regex_capturing = r"(la)+"
regex_noncapturing = r"(?:la)+"


print(re.findall(regex_capturing, test_string))
print(re.findall(regex_noncapturing, test_string))
````

### L'oggetto `Match`

Fino ad ora abbiamo solo stampato i risultati di `re.search`, `re.match`, `re.finditer` senza preoccuparci troppo di come funzionino gli oggetti `Match` che esse ritornano. Vediamo ora come usare questi oggetti, che rappresentano le occorrenze della regex all'interno della stringa.

Gli oggetti match ci consentono di accedere ai capturing group che abbiamo appena trovato tramite il metodo `m.group(i)` dove m è un `Match` e `i` è un numero intero. Il gruppo 0 rappresenta sempre l'intera stringa, i gruppi da 1 in poi sono quelli "veri".

````python
date_str = "02/01/1999"
date_regex = r"([0-9]{2})/([0-9]{2})/([0-9]{4})"

match = re.search(date_regex,date_str)
print(match.group(0))
print(match.group(1)) # day
print(match.group(2)) # month
print(match.group(3)) # year
````

Possiamo poi ottenere le informazioni sugli indici dei gruppi appena trovati nella stringa usando `m.start(i)` e `m.end(i)`, che ritornano gli indici di inizio e fine dell'i-esimo gruppo. Il valore `i` è opzionale, se non viene fornito si usa 0.

````python
date_str = "On the 02/01/1999 I went for a walk"
date_regex = r"([0-9]{2})/([0-9]{2})/([0-9]{4})"

match = re.search(date_regex,date_str)
print(match.start())
print(date_str[:match.start(2) - 1])
print(date_str[:match.end(1)])
````

### Il carattere `|`

L'ultimo carattere che vediamo oggi è `|`, che consente di fare match di una alternativa fra due stringhe. È un po' come l'or logico: una delle due stringhe va bene. 

````python
import re
test_regex = r"one|of|these"
test_string = "one two three is used to count the number of things"

print(re.findall(test_regex, test_string))
````

Bisogna usarlo con attenzione, perché senza gruppi considera tutte le stringhe alla sua sinistra e alla sua destra.

````python
test_string = "the beautiful coast"
test_regex = r"the beautiful mountain|coast"
test_regex_group = r"the beautiful (?:mountain|coast)"


print(re.search(test_regex, test_string))
print(re.search(test_regex_group, test_string))
````
